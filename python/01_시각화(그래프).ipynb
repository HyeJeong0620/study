```python
# 시각화 모듈
import matplotlib.pyplot as plt
```

```python
#부산 3월 기온
temp = [10.8, 8.5, 8.7, 7.9, 5.1, 
        8.1, 8.2, 10.4, 11, 9.9, 
        7.8, 9.5, 11.6, 8.2, 7.7, 
        6.6, 11.1, 12.2, 12.6, 10.4, 
        13.7, 14.1, 12.6, 11.5, 13, 
        14.5, 13.7, 9.4, 10.4, 11.6, 12.9]
```

```python
# 그래프 그리기
plt.plot(temp)
plt.show()
```

```python
# STEP 3. matplotlib의 폰트를 Nanum 폰트로 지정

plt.rc('font', family='NanumBarunGothic')
```

```python
# 제목 달기

plt.title('3월 기온')

plt.plot(temp)
plt.show()
```

```python
# 선옵션
plt.plot(temp, 'ro:')
plt.show()
```

```python
# 그래프 크기 조절: 가로 세로를 튜플로
plt.figure(figsize=(10,6))

# 그래프 제목 달기
plt.title('3월 기온')
plt.xlabel('날짜')
plt.ylabel('기온')

plt.plot(temp, 'go--', label='기본')
plt.plot(temp, 'b-', drawstyle='steps-post', label='계단식')
plt.legend(loc='best') # 범례

plt.grid(True, color='gray', alpha=0.5, linestyle='--')
plt.axhline(y=10, color='r', linewidth=1)

plt.show()
```

```python
# 여러 그래프 그리기
# 그래프 그리기
x = list(range(1,32)) # x축 날짜
plt.figure(figsize=(10,6))

plt.subplot(2,2,1) # 2행 2열 1번째(전체 4개 중 1번째)
plt.plot(x,temp,'go--')
plt.title('기본그래프')

plt.subplot(2,2,2) # 2행 2열 2번째(전체 4개 중 2번째)
plt.scatter(x,temp)
plt.title('산포도그래프')

plt.subplot(2,2,3) # 2행 2열 3번째(전체 4개 중 3번째)
plt.bar(x,temp)
plt.title('세로막대그래프')

plt.subplot(2,2,4) # 2행 2열 4번째(전체 4개 중 4번째)
plt.barh(x,temp) # 가로 막대그래프
plt.title('가로막대그래프')
```
